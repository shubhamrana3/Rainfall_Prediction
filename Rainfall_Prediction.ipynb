```python
#import libraries
import numpy as np
import pandas as pd
import sklearn as sk
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("austin-weater-final.csv")
```

```python
data
```

```python
#precipitation column will serve as label
x = data.drop(['PrecipitationSumInches'], axis=1)
y = data ["PrecipitationSumInches"]
```

```python
#reshaping it into 2nd vector
y = y.values.reshape(-1,1)
```

```python
y
```

```python
day_index=798
days = [i for i in range(y.size)]
days
```

```python
#initialise the Linear Regression Classifier
clf = LinearRegression()
```

```python
#train the classifier
clf.fit(x,y)
```

```python
#plot a graph
print ("The precipitation trend graph:")
plt.scatter(days, y, color='g')
plt.scatter(days[day_index], y[day_index], color='r')
plt.title("precipitation level")
plt.xlabel("Days")
plt.ylabel("Precipitation in inches")
plt.show()
x_vis = x.filter(['TempAvgF', 'DewPointAvgF', 'HumidityAvgPercent', 'SeaLevelPressureAvgInches', 'VisibilityAvgMiles', 'WindAvgMPH'])
```

```python
print("The Precipitation VS Attributes trend Graph:")
for i in range (x_vis.columns.size):
  plt.subplot(3,2,i+1)

  plt.scatter(days,x_vis[x_vis.columns.values[i][:100]], color = 'g')

  plt.scatter(days[day_index],x_vis[x_vis.columns.values[i]][day_index], color = 'r')

  plt.title(x_vis.columns.values[i])

plt.show()
```

```python
#conclusion:
we have one day in red, where the precipitation is about 2 inches.
so we can observe that the rainfall can be expected to be high when the temperature is high and the humidity is high.
```
